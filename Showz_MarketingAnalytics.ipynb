# Paso 1. Accediendo a los datos y preparandolos para el an√°lisis

### COMENTARIO:
Saber que tipo de datos tengo es importante para no tener problemas mas adelante al trabajar los datos.

```python
#Para el dateframe visits
visits['Start Ts'] = pd.to_datetime(visits['Start Ts'])
visits['End Ts'] = pd.to_datetime(visits['End Ts'])

#Para el dataframe orders
orders['Buy Ts'] = pd.to_datetime(orders['Buy Ts'])

#Para el dataframe costs
costs['dt'] = pd.to_datetime(costs['dt'])

#verificar el cambio 
print(visits.info())
print(orders.info())
print(costs.info())
```

# Paso 2. Informe y calculos de m√©tricas

## 1.Visitas

```python

# Extraer fechas, semana y mes
visits['date'] = visits['Start Ts'].dt.date
visits['week'] = visits['Start Ts'].dt.to_period('W').apply(lambda r: r.start_time.date())
visits['month'] = visits['Start Ts'].dt.to_period('M').apply(lambda r: r.start_time.date())

# Usuarios √∫nicos
daily_users = visits.groupby('date')['Uid'].nunique()
weekly_users = visits.groupby('week')['Uid'].nunique()
monthly_users = visits.groupby('month')['Uid'].nunique()

# Sesiones por d√≠a
sessions_per_day = visits.groupby('date').size()

# Duraci√≥n de sesiones en minutos
visits['session_duration'] = (visits['End Ts'] - visits['Start Ts']).dt.total_seconds() / 60

# Estad√≠sticas de duraci√≥n de sesiones
session_duration_stats = visits['session_duration'].describe()

# Frecuencia de retorno de usuarios
user_sessions = visits.groupby('Uid').agg(
    sessions_count=('Start Ts', 'count'),
    active_days=('date', 'nunique')
)

repeat_users_percentage = (user_sessions['sessions_count'] > 1).mean() * 100
returning_users_percentage = (user_sessions['active_days'] > 1).mean() * 100

print(f"Porcentaje de usuarios que vuelven (m√°s de 1 sesi√≥n): {repeat_users_percentage:.2f}%")
print(f"Porcentaje de usuarios que visitan en m√°s de un d√≠a: {returning_users_percentage:.2f}%")

```

### COMENTARIO: 
El 22.85% de los usuarios vuelve y tiene m√°s de una sesi√≥n en Showz. Adem√°s, el 19.74% visita en m√°s de un d√≠a diferente. Esto muestra que casi una de cada cinco personas usa la plataforma varias veces, lo que es bueno para la retenci√≥n.

## 2.Ventas

```python
# 1. Fecha de primera visita y primera compra por usuario
first_visit = visits.groupby('Uid')['Start Ts'].min().reset_index().rename(columns={'Start Ts': 'first_visit_date'})
first_purchase = orders.groupby('Uid')['Buy Ts'].min().reset_index().rename(columns={'Buy Ts': 'first_purchase_date'})

# Unir datos de visitas y compras para calcular tiempo hasta la primera compra
user_conversion = pd.merge(first_visit, first_purchase, on='Uid', how='left')
user_conversion['days_to_purchase'] = (user_conversion['first_purchase_date'] - user_conversion['first_visit_date']).dt.days

# Clasificar conversiones (usuarios que no compraron tendr√°n -1)
user_conversion['conversion_bucket'] = user_conversion['days_to_purchase'].fillna(-1).astype(int)

# Agregar cohorte (mes de primera visita)
user_conversion['cohort_month'] = user_conversion['first_visit_date'].dt.to_period('M').astype(str)

# Filtrar datos para d√≠as entre 0 y 30 (solo usuarios que compraron)
heatmap_data = user_conversion[(user_conversion['days_to_purchase'] >= 0) & (user_conversion['days_to_purchase'] <= 30)]

# Crear tabla pivot para heatmap: filas = cohortes, columnas = d√≠as hasta compra, valores = conteo usuarios
pivot = heatmap_data.pivot_table(index='cohort_month', columns='days_to_purchase', values='Uid', aggfunc='count').fillna(0)

# Mapa de calor para las conversiones
plt.figure(figsize=(16,10))
sns.heatmap(pivot, cmap='YlGnBu', linewidths=.5)
plt.title('Heatmap: N√∫mero de usuarios por cohorte y d√≠as hasta primera compra')
plt.xlabel('D√≠as hasta primera compra')
plt.ylabel('Mes de primera visita (cohorte)')
plt.show()

# ------------------ Otras m√©tricas y gr√°ficos de ventas ------------------

# 2. Pedidos por mes (Gr√°fico de l√≠neas)
orders['month'] = orders['Buy Ts'].dt.to_period('M')
orders_per_month = orders.groupby('month').size().reset_index(name='orders_per_month')
orders_per_month['month'] = orders_per_month['month'].astype(str)

plt.figure(figsize=(12,6))
plt.plot(orders_per_month['month'], orders_per_month['orders_per_month'], marker='o')
plt.xticks(rotation=45)
plt.title('N√∫mero de pedidos por mes')
plt.xlabel('Mes')
plt.ylabel('N√∫mero de pedidos')
plt.grid(True)
plt.tight_layout()
plt.show()

# 3. Tama√±o promedio de compra (Gr√°fico de barras)
average_order_size = orders['Revenue'].mean()
print(f"Tama√±o promedio de compra: {average_order_size:.2f}")

# Histograma del tama√±o de compra (convertido a gr√°fico de barras)
plt.figure(figsize=(10,6))
plt.bar(orders['Revenue'].value_counts().index, orders['Revenue'].value_counts(), width=10, alpha=0.7)
plt.title('Distribuci√≥n del tama√±o de compra (Revenue por pedido)')
plt.xlabel('Revenue')
plt.ylabel('N√∫mero de pedidos')
plt.grid(True)
plt.show()

# 4. LTV por usuario (Gr√°fico de barras)
ltv_per_user = orders.groupby('Uid')['Revenue'].sum().reset_index().rename(columns={'Revenue': 'ltv'})

# Histograma del LTV (convertido a gr√°fico de barras)
plt.figure(figsize=(10,6))
plt.bar(ltv_per_user['ltv'].value_counts().index, ltv_per_user['ltv'].value_counts(), width=10, alpha=0.7)
plt.title('Distribuci√≥n del LTV (Ingresos totales por usuario)')
plt.xlabel('LTV')
plt.ylabel('N√∫mero de usuarios')
plt.grid(True)
plt.show()



```

### 1. ¬øCu√°nto dinero se gast√≥? (Total/por fuente de adquisici√≥n/a lo largo del tiempo)

```python

# Calcular gasto total
total_cost = costs['costs'].sum()
print(f"Gasto total en marketing: {total_cost:.2f}")

# Gasto por fuente de adquisici√≥n
costs_by_source = costs.groupby('source_id')['costs'].sum().reset_index()
costs_by_source = costs_by_source.rename(columns={'costs': 'total_cost'})
print(costs_by_source)

# Gasto a lo largo del tiempo (por mes)
costs['month'] = costs['dt'].dt.to_period('M')
monthly_costs = costs.groupby('month')['costs'].sum().reset_index()
print(monthly_costs)

# Gr√°fico de barras: Gasto por fuente de adquisici√≥n
plt.figure(figsize=(10,6))
plt.bar(costs_by_source['source_id'].astype(str), costs_by_source['total_cost'])
plt.title('Gasto Total en Marketing por Fuente de Adquisici√≥n')
plt.xlabel('Fuente de Adquisici√≥n')
plt.ylabel('Gasto Total')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Gr√°fico de barras: Gasto mensual a lo largo del tiempo
plt.figure(figsize=(12,6))
plt.bar(monthly_costs['month'].astype(str), monthly_costs['costs'])
plt.title('Gasto Mensual en Marketing a lo Largo del Tiempo')
plt.xlabel('Mes')
plt.ylabel('Gasto Total')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

###  2. ¬øCu√°l fue el costo de adquisici√≥n de clientes de cada una de las fuentes?
El Costo de Adquisici√≥n de Clientes se calcula como el gasto total de marketing dividido entre el n√∫mero de nuevos clientes adquiridos

```python
# Paso 1: Sumar los costos por fuente (total por 'source_id' en `costs`)
costs_by_source = costs.groupby('source_id')['costs'].sum().reset_index()
costs_by_source = costs_by_source.rename(columns={'costs': 'total_cost'})  # Renombrar la columna de costos

# Paso 2: Contar los usuarios √∫nicos por cohorte (mes de la primera visita)
clients_by_cohort = user_conversion.groupby('cohort_month')['Uid'].nunique().reset_index()
clients_by_cohort = clients_by_cohort.rename(columns={'Uid': 'num_clients'})  # Renombrar para mayor claridad

# Paso 3: Aqu√≠ necesitamos una suposici√≥n: Los clientes adquiridos en cada cohorte pueden estar relacionados con las fuentes de adquisici√≥n
# Suponemos que cada cohorte tiene la misma cantidad de fuentes distribuidas. Para esto, necesitamos hacer la distribuci√≥n correcta.
sources = costs_by_source['source_id'].unique()  # Obtener las fuentes disponibles
num_sources = len(sources)
num_cohorts = len(clients_by_cohort)

# Asignar las fuentes de adquisici√≥n a las cohortes. Vamos a distribuirlas equitativamente.
# Repetimos las fuentes hasta cubrir todas las cohortes
sources_repeated = np.tile(sources, int(np.ceil(num_cohorts / num_sources)))[:num_cohorts]  # Repetir hasta que cubra todas las cohortes

# Asignamos las fuentes repetidas a las cohortes
clients_by_cohort['source_id'] = sources_repeated

# Paso 4: Unir los datos de costos con los clientes por fuente
result = pd.merge(costs_by_source, clients_by_cohort, on='source_id', how='left')

# Paso 5: Calcular el Costo de Adquisici√≥n de Clientes (CAC) por fuente
result['cac'] = result['total_cost'] / result['num_clients']

# Paso 6: Mostrar el resultado final
print(result)



```

### COMENTARIO 
El costo de adquisici√≥n de clientes (CAC) vari√≥ seg√∫n la fuente y el mes. Las fuentes 9 y 10 tuvieron el CAC m√°s bajo (0.20 y 0.23), siendo las m√°s eficientes. En cambio, la fuente 3 present√≥ el CAC m√°s alto, llegando a 13.88 en un mes. Otras fuentes mostraron costos intermedios que fluctuaron con el tiempo. Esto indica que algunas fuentes son m√°s rentables para conseguir clientes que otras.

```python
# Paso 1: Fusionar los DataFrames 'orders' y 'costs' para obtener 'source_id' en 'orders'
orders_with_source = pd.merge(orders, costs[['month', 'source_id']], on='month', how='left')

# Paso 2: Fusionar el DataFrame resultante con 'user_conversion' para obtener 'cohort_month' (si es necesario)
orders_with_source_and_conversion = pd.merge(orders_with_source, user_conversion[['Uid', 'cohort_month']], on='Uid', how='left')

# Verificar las columnas del DataFrame resultante
print(orders_with_source_and_conversion.columns)

# Paso 3: Sumar los ingresos generados por cada fuente (source_id)
orders_per_source = orders_with_source_and_conversion.groupby('source_id')['Revenue'].sum().reset_index()
orders_per_source = orders_per_source.rename(columns={'Revenue': 'total_revenue'})  # Renombrar para mayor claridad

# Paso 4: Sumar los costos de marketing por fuente (source_id)
costs_by_source = costs.groupby('source_id')['costs'].sum().reset_index()
costs_by_source = costs_by_source.rename(columns={'costs': 'total_cost'})  # Renombrar para mayor claridad

# Paso 5: Unir los ingresos y los costos por fuente
romi_data = pd.merge(costs_by_source, orders_per_source, on='source_id', how='left')

# Paso 6: Calcular el ROMI por fuente
romi_data['romi'] = (romi_data['total_revenue'] - romi_data['total_cost']) / romi_data['total_cost']

# Paso 7: Mostrar el resultado final con el ROMI por fuente
print(romi_data[['source_id', 'romi']])
```

```python
# Mostrar las primeras filas de 'visits' para inspeccionar las columnas y verificar si 'Source Id' existe
print("Primeras filas de 'visits':")
print(visits.head())

# Mostrar las columnas de 'visits' para verificar la existencia de 'Source Id'
print("\nColumnas de 'visits' antes de renombrar:")
print(visits.columns)

# Asegur√©monos de que 'Source Id' se renombre correctamente a 'source_id'
# Si la columna 'Source Id' est√° en 'visits', renombr√©mosla a 'source_id'
if 'Source Id' in visits.columns:
    visits.rename(columns={'Source Id': 'source_id'}, inplace=True)

# Verificar si el renombrado ha sido exitoso
print("\nColumnas de 'visits' despu√©s de renombrar 'Source Id' a 'source_id':")
print(visits.columns)

# Verificar si 'source_id' ahora est√° presente en 'visits'
print("\nVerificando 'source_id' en 'visits':")
print(visits['source_id'].head())
```

```python

# Asegur√©monos de que las columnas est√©n correctamente renombradas y presentes

# 1. Fusionar 'orders' con 'visits' para agregar el 'source_id'
orders_with_source = pd.merge(orders, visits[['Uid', 'source_id']], on='Uid', how='left')

# Verificar si la fusi√≥n fue exitosa
print("\nResultado de la fusi√≥n entre 'orders' y 'visits' (con 'source_id'):")
print(orders_with_source.head())

# 2. Fusionar 'orders_with_source' con 'costs' para obtener los costos y la fecha
orders_with_source_and_conversion = pd.merge(orders_with_source, costs[['source_id', 'dt', 'costs']], on='source_id', how='left')

# Verificar las columnas despu√©s de la fusi√≥n
print("\nResultado de la fusi√≥n entre 'orders_with_source' y 'costs':")
print(orders_with_source_and_conversion.head())

# 3. Agregar la columna 'month' en 'orders_with_source_and_conversion' para que podamos agrupar por mes
orders_with_source_and_conversion['month'] = orders_with_source_and_conversion['dt'].dt.to_period('M')

# Verificar si la columna 'month' se agreg√≥ correctamente
print("\nPrimeras filas con la columna 'month' agregada:")
print(orders_with_source_and_conversion.head())

# 4. Calcular el ROMI: (Revenue - Costos) / Costos
orders_with_source_and_conversion['romi'] = (orders_with_source_and_conversion['Revenue'] - orders_with_source_and_conversion['costs']) / orders_with_source_and_conversion['costs']

# Verificar las primeras filas con el ROMI calculado
print("\nPrimeras filas con el ROMI calculado:")
print(orders_with_source_and_conversion[['month', 'source_id', 'romi']].head())

# 5. Graficar el ROMI por fuente de adquisici√≥n
plt.figure(figsize=(12,6))

# Agrupar por 'month' y 'source_id' para calcular el ROMI promedio mensual
monthly_romi = orders_with_source_and_conversion.groupby(['month', 'source_id']).agg({'Revenue': 'sum', 'costs': 'sum'}).reset_index()

# Calcular el ROMI
monthly_romi['romi'] = (monthly_romi['Revenue'] - monthly_romi['costs']) / monthly_romi['costs']

# Trazar el gr√°fico de l√≠neas
for source in monthly_romi['source_id'].unique():
    source_data = monthly_romi[monthly_romi['source_id'] == source]
    plt.plot(source_data['month'].astype(str), source_data['romi'], label=f"Fuente {source}")

# Etiquetas y t√≠tulo
plt.xlabel('Mes')
plt.ylabel('ROMI')
plt.title('Evoluci√≥n del ROMI por Fuente de Adquisici√≥n a lo largo del Tiempo')
plt.legend(title="Fuente de Adquisici√≥n")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```
